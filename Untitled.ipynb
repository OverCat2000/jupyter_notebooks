## education

```python
lr50 = df.groupby(["income", "education"])["education"].count().loc[" <=50K"]
```

```python
ur50 = df.groupby(["income", "education"])["education"].count().loc[" >50K"]
```

```python
edu = pd.merge(lr50, ur50, right_index=True, left_index=True)
```

```python
edu.columns = ["lr50", "ur50"]
```

```python
edu[edu.lr50 < edu.ur50]
```

```python
edu["total"] = edu.lr50 + edu.ur50
```

```python
edu.lr50 = edu.lr50/edu.total
edu.ur50 = edu.ur50/edu.total
```

```python
edu.drop(["total"], axis=1, inplace=True)
```

```python
edu.sort_values(by="ur50", ascending=True, inplace=True)
```

```python
edu.plot(kind="barh",
         stacked=True,
         colormap='tab10')
plt.ylabel("count")
L = plt.legend(loc=1)
L.get_texts()[0].set_text(" <=50K")
L.get_texts()[1].set_text(" >50K")
plt.savefig("edupercentage.jpg", bbox_inches='tight')
```

## Occupation

```python
df.loc[df["occupation"] == " ?", "occupation"] = "unkown"
```

```python
occu = pd.crosstab(index=df["income"],
            columns=df["occupation"])
```

```python
occu = occu.T
```

```python
occu
```

```python
occu["l50"] = occu.apply(lambda x: x[" <=50K"] / (x[" <=50K"] + x[" >50K"])*100, axis=1)
occu["u50"] = occu.apply(lambda x: x[" >50K"] / (x[" <=50K"] + x[" >50K"])*100, axis=1)
```

```python
occu
```

```python
occu.sort_values(by="u50", ascending=True, inplace=True)
```

```python
occu.iloc[:, 2:4].plot(kind="bar",
         stacked=True,
         colormap='tab10')
plt.ylabel("count")
L = plt.legend()
L.get_texts()[0].set_text(" <=50K")
L.get_texts()[1].set_text(" >50K")
```

```python
df.loc[df["workclass"] == " ?", "workclass"] = "unknown"
```

```python
work = pd.crosstab(index=df["workclass"],
                   columns=df["income"])
```

```python
work["l50"] = work.apply(lambda x: x[" <=50K"] / (x[" <=50K"] + x[" >50K"])*100, axis=1)
work["u50"] = work.apply(lambda x: x[" >50K"] / (x[" <=50K"] + x[" >50K"])*100, axis=1)
```

```python
work.iloc[:, 2:4].plot(kind="bar", 
                       stacked=True,
                       colormap="tab10")
```

```python
def stackedbar(var, df):
    temp = pd.crosstab(index=df[var],
                       columns=df["income"])
    temp["l50"] = temp.apply(lambda x: x[" <=50K"] / (x[" <=50K"] + x[" >50K"])*100, axis=1)
    temp["u50"] = temp.apply(lambda x: x[" >50K"] / (x[" <=50K"] + x[" >50K"])*100, axis=1)

    temp.sort_values(by="u50", ascending=True, inplace=True)
    temp.iloc[:, 2:4].plot(kind="bar",
                           stacked=True,
                           colormap="tab10")
                
```
